# Are there any unique trends in the education data?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_2016 = pd.read_csv('~/Downloads/c2016_a.csv')
data_2015 = pd.read_csv('~/Downloads/c2015_a.csv')
data_2014 = pd.read_csv('~/Downloads/C2014_A/c2014_a.csv')
data_2013 = pd.read_csv('~/Downloads/C2013_A/c2013_a.csv')
data_2012 = pd.read_csv('~/Downloads/C2012_A/c2012_a.csv')
data_2011 = pd.read_csv('~/Downloads/C2011_A/c2011_a.csv')
data_2010 = pd.read_csv('~/Downloads/C2010_A/c2010_a.csv')
```

```python
data_2012.dropna(axis=1, how='all')
data_2011.dropna(axis=1, how='all')
data_2010.dropna(axis=1, how='all')
```

# Comparison between the code associated with degrees(CIPCODE) and the number of total women and men students in a school(CTOTALW, CTOTALM). 

```python
pal = dict(Lunch="seagreen", Dinner="gray")
g = sns.FacetGrid(data_2016, palette=pal, size=5)
g.map(plt.scatter, 'CIPCODE', 'CTOTALW', s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.map(plt.scatter, 'CIPCODE', 'CTOTALM', s=50, alpha=.7, linewidth=.5, color='r')
g.add_legend()
g.set(xlim=(0, 55))
g.set(ylim=(0,6000))

g = sns.FacetGrid(data_2010, palette=pal, size=5)
g.map(plt.scatter, 'CIPCODE', 'CTOTALW', s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.map(plt.scatter, 'CIPCODE', 'CTOTALM', s=50, alpha=.7, linewidth=.5, color='r')
g.add_legend()
g.set(xlim=(0, 55))
g.set(ylim=(0,6000))
```

# Comparison between the code associated with the number of degrees awarded and the number of total women and men students in a school(CTOTALW, CTOTALM). 

```python
pal = dict(Lunch="seagreen", Dinner="gray")
g = sns.FacetGrid(data_2016, palette=pal, size=5)
g.map(plt.scatter, 'MAJORNUM', 'CTOTALW', s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.map(plt.scatter, 'MAJORNUM', 'CTOTALM', s=50, alpha=.7, linewidth=.5, color='r')
g.add_legend()
g.set(xlim=(0, 3))
g.set(ylim=(0,25000))

g = sns.FacetGrid(data_2010, palette=pal, size=5)
g.map(plt.scatter, 'MAJORNUM', 'CTOTALW', s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.map(plt.scatter, 'MAJORNUM', 'CTOTALM', s=50, alpha=.7, linewidth=.5, color='r')
g.add_legend()
g.set(xlim=(0,3))
g.set(ylim=(0,25000))

```
